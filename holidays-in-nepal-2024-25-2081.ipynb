```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.qppstudio.net/publicholidays2024/nepal.htm"
```

```python
content = requests.get(url)
```

```python
soup = BeautifulSoup(content.text, 'html')
```

```python
table = soup.find("table")
```

```python
headers = table.find_all("th")
headers
```

```python
titles = []
titles
```

```python
for i in headers:
    title = i.text
    titles.append(title)
```

```python
titles
```

```python
df = pd.DataFrame(columns = titles)
df
```

```python
rows = table.find_all("tr")
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]
    l = len(df)
    df.loc[l] = row
```

```python
df
```

```python
df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
```

```python
df['Public Holiday'] = df['Public Holiday'].str.replace('\r\n', '')
```

```python
df
```

```python
df.to_csv("Public Holidays in Nepal for the year 2081 BS.csv", index = False)
```
